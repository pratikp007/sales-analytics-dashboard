```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('train.csv', encoding='latin-1')
print('Data loaded!', df.shape)
```

```
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print(category_sales)

plt.figure(figsize=(8,5))
category_sales.plot(kind='bar', color=['#2E86C1','#E67E22','#1E8449'])
plt.title('Total Sales by Category', fontsize=14, fontweight='bold')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('sales_by_category.png', dpi=150)
plt.show()
print('Chart 1 done!')
```

```
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print(region_sales)

plt.figure(figsize=(8,5))
region_sales.plot(kind='bar', color=['#1A3A5C','#2E86C1','#5DADE2','#AED6F1'])
plt.title('Total Sales by Region', fontsize=14, fontweight='bold')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('sales_by_region.png', dpi=150)
plt.show()
print('Chart 2 done!')
```

```
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
print(top_customers)

plt.figure(figsize=(10,6))
top_customers.plot(kind='barh', color='#2E86C1')
plt.title('Top 10 Customers by Sales', fontsize=14, fontweight='bold')
plt.xlabel('Total Sales')
plt.tight_layout()
plt.savefig('top_customers.png', dpi=150)
plt.show()
print('Chart 3 done!')
```

```
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Month'] = df['Order Date'].dt.month

monthly = df.groupby('Month')['Sales'].sum()

plt.figure(figsize=(10,5))
plt.plot(monthly.index, monthly.values, marker='o', linewidth=2, color='#2E86C1')
plt.fill_between(monthly.index, monthly.values, alpha=0.15, color='#2E86C1')
plt.title('Monthly Sales Trend', fontsize=14, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(range(1,13), ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.tight_layout()
plt.savefig('monthly_trend.png', dpi=150)
plt.show()
print('Chart 4 done!')
```
